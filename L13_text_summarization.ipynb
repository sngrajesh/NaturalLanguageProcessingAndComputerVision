````python
import urllib
from urllib.request import urlopen
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize

````

````python
url = urlopen('https://en.wikipedia.org/wiki/Rajgad_Fort')
html = url.read()
````

````python
html
````

````python
soup = BeautifulSoup(html, 'html.parser')
soup 
````

````python
soup.find('title').text
````

````python
data = soup.get_text(strip=True)
print(data)
````

````python
for tag in soup.find_all('p'):
    print(tag.text)
````

````python

tokens = word_tokenize(data)
tokens
````

````python
from nltk import FreqDist
````

````python
freq = FreqDist(tokens)
````

````python
freq.plot(20)
````

````python
from nltk.corpus import stopwords
````

````python
tokens1 = [token for token in tokens if token.isalpha() or token.isdigit()]
swords = stopwords.words('english')
tokens2 = [word.lower() for word in tokens1 if word.lower() not in swords]
````

````python
freq = FreqDist(tokens2)
freq.plot(20)
````

````python
freq.most_common(5)
````

````python
url = urlopen('https://en.wikipedia.org/wiki/Rajgad_Fort')
html = url.read()
soup = BeautifulSoup(html, 'html.parser')
data = soup.get_text()
tokens = word_tokenize(data)
swords = stopwords.words('english')
tokens1 = [token for token in tokens if token.lower() not in swords]
tokens2 = [token.lower() for token in tokens1 if token.isalnum()]
freq = FreqDist(tokens2)
freq.plot(20)
freq.most_common(5)
````

````python
from nltk.stem import PorterStemmer

ls = PorterStemmer()

url = urlopen('https://en.wikipedia.org/wiki/Tiger')
html = url.read()
soup = BeautifulSoup(html, 'html.parser')
data = soup.get_text()
tokens = word_tokenize(data)
swords = stopwords.words('english')
tokens1 = [token for token in tokens if token.lower() not in swords]
tokens2 = [token.lower() for token in tokens1 if token.isalnum()]
tokens3 = [ls.stem(token) for token in tokens2 if len(token)> 1]
freq = FreqDist(tokens3)
freq.plot(20)
freq.most_common(5)
````

````python
with open('Dataset/course.txt', 'r+', encoding='utf8') as file:
    ls = PorterStemmer()
    data = file.read()
    tokens = word_tokenize(data)
    swords = stopwords.words('english')
    tokens1 = [token for token in tokens if token.lower() not in swords]
    tokens2 = [token.lower() for token in tokens1 if token.isalnum()]
    tokens3 = [ls.stem(token) for token in tokens2 if len(token)> 1]
    freq = FreqDist(tokens3)
    freq.plot(20)
    print(freq.most_common(5))
````
