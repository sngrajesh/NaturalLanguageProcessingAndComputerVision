# Q.1

sent = 'They told that their ages are 23, 25 and 29 respectively.'
* Find the average of ages.
* Count total number of words starts with 't' / 'T'
* Count total punctuation symbols.
* Print the words in capital letter which ends with 'd'
* Merge all the words and print (no punctuations)


```python
import re
from string import punctuation
```

```python
sent = 'They told that their ages are 23, 25 and 29 respectively.'
```

```python
# Find the average of the age.
ages = re.findall(r'\d+', sent)
ages = [int(age) for age in ages]
print(sum(ages)/len(ages))
```

```python
# COunt the number of words starts with t/T.
print(sum([1 for word in sent.split() if word.startswith('t')]))
```

```python
# count total punnctuation symbols.
print(len(set([x for x in sent if x in punctuation])))
```

```python
# Print the words in capital letter which ends with 'd'
print([word for word in sent.split() if word.endswith('d') and word[0].isupper()])
```

```python
# Merge all the words and print (no punctuations)
print(' '.join([word for word in sent.split() if word.isalpha() and word not in punctuation]))
```

# Q.2

Open the wikipedia page of Indian flag and perform following analysis
* Count total number of Words
* Count percentage of punctuation marks
* Print total number of lines
* How many words are in title case?
* Remove the numbers and print the text

```python
import re
import requests
from bs4 import BeautifulSoup 
from nltk.tokenize import word_tokenize, sent_tokenize
from string import punctuation
```

```python
url = r'https://en.wikipedia.org/wiki/Flag_of_India'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
text = soup.get_text()
```

```python
total_text = word_tokenize(text)
print('Count total number of Words : ',len(total_text))
```

```python
token1 = [word for word in word_tokenize(text) if word in punctuation]
print(len(token1)/len(text)*100)
```

```python
total_lines = 0
for line in text.split('\n'):
    total_lines += 1
print('Print total number of lines : ',total_lines)
```

```python
total_title_case = 0
for word in total_text:
    if word.istitle():
        total_title_case += 1
print('How many words are in title case? : ',total_title_case)
```

```python
pattenr = re.compile(r'\d+')
text_without_num = pattenr.sub('', text)
print('Remove the numbers and print the text : ',text_without_num)
```
