```python
from flask import Flask, url_for, redirect, request, render_template_string, render_template
app = Flask(__name__)


def get_hello_html(text):
    return f'''
        <html><body><h3>
            {text}
        </h3></body></html>
    '''

def get_form_template(page):
    return f'''
        <h2>{page}</h2>
        <a href='/'>Home</a> 
        <a href='test'>Test</a><hr>
        <form action="/" method="post">
            <input type="text" name="name" placeholder="Enter your name">
            <input type="submit" value="Submit">
        </form>
    '''

@app.route('/test/', defaults={'user': None})
@app.route('/test/<user>')
def test(user):
    if not user:
        test_page = get_form_template('Test')
        return render_template_string(test_page)
    
    return render_template('test.html', user=user)


@app.route('/admin')
def hello_admin():
    return get_hello_html(f'Welcome Admin!!!')


@app.route('/guest/<guest>')
def hello_guest(guest):
    return get_hello_html(f'Welcome {guest}')


@app.route('/user/<name>')
def hello(name):
    if name == 'admin':
        return redirect(url_for('hello_admin'))
    else:
        return redirect(url_for('hello_guest', guest=name))


@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'POST':
        name = request.form['name']
        return redirect(url_for('hello', name=name))
    home_page = get_form_template('Home')
    return render_template_string(home_page) 

if __name__ == '__main__':
    app.run()
```

```python
from flask import Flask, url_for, redirect, request, render_template_string, render_template

app = Flask(__name__)

def get_html_skeleton(children):
    return f"""
        <!DOCTYPE html>
        <html lang="en">
        <head>
            <meta charset="UTF-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Test</title>
        </head>
        <body>
            <a href='/'>Home</a><hr>
            {children}
        </body>
        </html>
    """

def get_home_template():
    return '''
        <h2>Home</h2>
        <form action="/" method="post">
            <input type="text" name="name" placeholder="Enter your name"><br>
            <input type="number" min="0" max="100" name="physics" placeholder="Enter Physics marks :"><br>
            <input type="number" min="0" max="100" name="chemistry" placeholder="Enter Chemistry marks :"><br>
            <input type="number" min="0" max="100" name="mathmatics" placeholder="Enter Mathmatics marks :"><br>
            <input type="submit" value="Submit">
        </form>
    '''

def get_result_template():
    return '''       
        <h2>Result</h2>
        <table>
            <tr>
                <th>Name</th><th>Average</th>
            </tr>
            {% for name, average in results.items() %}
            <tr>
                <td>{{name}}</td><td>{{average}}</td>
            </tr>
            {% endfor %}
        </table>
    '''

@app.route('/result', methods=['GET'])
def show_result():
    result = request.args.get('results')
    result_dict = eval(result)
    result_page = get_html_skeleton(get_result_template())
    return render_template_string(result_page, results=result_dict)

results = {}

@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'POST':
        name = request.form['name']
        physics = int(request.form['physics'])
        chemistry = int(request.form['chemistry'])
        mathmatics = int(request.form['mathmatics'])
        average = (physics+chemistry+mathmatics)/3

        results[name] = average
        
        return redirect(url_for('show_result', results=results))

    home_page = get_html_skeleton(get_home_template())
    return render_template_string(home_page) 

if __name__ == '__main__':
    app.run()
```

```python
template = '''
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>Email Form</title>
        <style>
            body {
                font-family: Arial, sans-serif;
                background-color: #f4f4f4;
                display: flex;
                justify-content: center;
                align-items: center;
                height: 100vh;
                margin: 0;
            }
            .container {
                background-color: #fff;
                padding: 20px;
                border-radius: 8px;
                box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
                width: 600px;
            }
            label {
                display: block;
                margin-bottom: 8px;
                font-weight: bold;
            }
            textarea {
                width: 100%;
                height: 100px;
                padding: 10px;
                margin-bottom: 15px;
                border: 1px solid #ddd;
                border-radius: 4px;
                box-sizing: border-box;
                resize: vertical;
            }
            input[type="submit"] {
                background-color: #4CAF50;
                color: #fff;
                border: none;
                padding: 10px;
                border-radius: 4px;
                cursor: pointer;
                font-size: 16px;
            }
            input[type="submit"]:hover {
                background-color: #45a049;
            }
        </style>
    </head>
    <body>
        <div class="container">
            <form action="/" method="post">
                <label for="email">Email Message:</label>
                <textarea 
                    id="email" 
                    name="email"
                    placeholder="Type your message here..." 
                    required>{% if response.message %}{{ response.message }}{% endif %}</textarea>

                <input type="submit" value="Check">

                {% if response.result %}
                    <p>{{ response.result }}</p>
                {% endif %}
                
            </form>
        </div>
    </body>
    </html>
'''
```

```python
#*******************
# Import Libraries
#*******************

import joblib 
import warnings
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from flask import Flask, render_template_string , request




#*******************
# Prediction Logic
#*******************

warnings.filterwarnings('ignore')
swords = stopwords.words('english')
ps = PorterStemmer()
 
def clean_sentence(sent):
    tokens1 = word_tokenize(sent)
    tokens2 = [token for token in tokens1 if (token.isalpha() or token.isdigit())]
    tokens3 = [token.lower() for token in tokens2 if token.lower() not in swords ]
    tokens4 = [ps.stem(token) for token in tokens3]
    return tokens4

def check(message):
    classifier = joblib.load(r'models/mail_classifier.model')
    tfidf = joblib.load(r'models/mail_tfidf.model')

    message = clean_sentence(message)

    result = classifier.predict(tfidf.transform([' '.join(message)]))
    return f'Message is {str(result[0])}'




#*******************
# Actual Application
#*******************

app = Flask(__name__)
 
@app.route('/', methods=['GET', 'POST'], endpoint='index')
def index():
    response = {
            'message':None,
            'result':None
        }
    if request.method == 'POST':
        message = request.form.get('email')
        result = check(message=message)
        response = {
            'message':message,
            'result':result
        }
    return render_template_string(template, response=response)





#*******************
# Entry Point Here
#*******************

if __name__ == '__main__':
    app.run(debug=False)
```
