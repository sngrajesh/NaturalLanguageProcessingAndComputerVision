Theoretical Probability
It is based on the possible chances of something to happen.
The theoretical probability is mainly based on the reasoning behind probability.
For example, if a coint is tossed, the theoretical probability of getting a head will be 1/2.

Experimental Probability
It is based on the basis of the observations of an experiment.
The experimental probability can be calculated based on the number of possible outcomes by the total number of trials.




##### Theoretical Probability

```python

def coin_flip_probability():
  heads_probability = 1/2
  return heads_probability


def rolling_dice_probability(number):
  if number not in range(1,7):
    raise ValueError("Number must be between 1 and 6")
  probability = 1/6
  return probability
```

```python
coin_flip_result = coin_flip_probability()
dice_roll_result = rolling_dice_probability(4)

print('Probability of getting a Heads on a coin flip: ', coin_flip_result)
print('Probability of rolling a 4 on a dice: ', dice_roll_result)
```


##### Expreimental Probability

```python
import random
import numpy as np
import pandas as pd
from collections import Counter
```

```python

def simulate_coind_flip(numflips):
    result = []
    for _ in range(numflips):
        flip = random.choice(['Heads', 'Tails'])
        result.append(flip)
    return result
```

```python
data = simulate_coind_flip(1000) 
counts = Counter(data)

df = pd.Series(counts)
df.plot(kind='bar')
```

```python
def calculate_experimental_probability(results, event):
    event_count = results.count(event)
    total_count = len(results)
    probability = event_count / total_count
    return probability
```

```python
num_flips = 1000
simulation_results = simulate_coind_flip(num_flips)

head_probability = calculate_experimental_probability(simulation_results, "Heads")
tail_probability = calculate_experimental_probability(simulation_results, "Tails")

print(f"The number of flips is {num_flips}")
print(f"The probability of getting Heads is {head_probability:.2f}")
print(f"The probability of getting Tails is {tail_probability:.2f}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iq = np.array([
    70, 70, 80, 80, 80, 90, 90, 90, 90, 
    100, 100, 100, 100, 100, 
    110, 110, 110, 110, 
    120, 120, 120, 130, 130
])

# Plot probability density function
plt.figure(figsize=(6,4))
sns.kdeplot(iq, color='red', linewidth=2)
plt.xlabel('IQ')
plt.ylabel('Frequency')
plt.title('IQ Distribution')
plt.show()

display(iq.mean(),np.median(iq), pd.Series(iq).mode()[0])
```

```python
# Losgistics Growth
x =  np.arange(-5, 5, 0.25)

def sigmoid(z):
    val = 1 / (1 + np.exp(-z))
    return val

plt.plot(x, sigmoid(x))
```
